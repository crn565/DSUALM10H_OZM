## coeficientes de correlacion

Para calcular el coeficiente de correlación de Pearson entre las métricas y el número de muestras, puedes utilizar una herramienta estadística como Python y la biblioteca NumPy. Aquí tienes un ejemplo de cómo puedes hacerlo:
    Este código calculará el coeficiente de correlación de Pearson entre las muestras y cada una de las métricas. Los resultados se imprimirán en la consola.

```python
import numpy as np

# Datos de las muestras y las métricas
muestras = np.array([25801, 100081, 15664, 12544, 27443, 31460, 36047, 249040, 223234, 123154, 107491, 94948])
f1 = np.array([0.5265, 0.584, 0.4658, 0.4854, 0.5846, 0.5966, 0.5982, 0.5664, 0.565, 0.5672, 0.5678, 0.5798])
EAE = np.array([0.056, 0.023, 0.003, 0.0028, 0, 0, 0, 0, 0, 0, 0, 0])
MNEAP = np.array([0.7246, 1.0466, 1.08, 0.96, 0.903, 81.9176, 0.1014, 4.381, 7.1404, 2.9092, 2.7012, 2.4926])
RMSE = np.array([14.7822, 15.9096, 13.456, 12.5952, 12.214, 2.9964, 6.0592, 12.4878, 13.2298, 8.0798, 7.6894, 8.263])

# Cálculo del coeficiente de correlación de Pearson
correlation_samples_f1 = np.corrcoef(muestras, f1)[0, 1]
correlation_samples_EAE = np.corrcoef(muestras, EAE)[0, 1]
correlation_samples_MNEAP = np.corrcoef(muestras, MNEAP)[0, 1]
correlation_samples_RMSE = np.corrcoef(muestras, RMSE)[0, 1]

print("Coeficiente de correlación (muestras - f1):", correlation_samples_f1)
print("Coeficiente de correlación (muestras - EAE):", correlation_samples_EAE)
print("Coeficiente de correlación (muestras - MNEAP):", correlation_samples_MNEAP)
print("Coeficiente de correlación (muestras - RMSE):", correlation_samples_RMSE)
```

 El coeficiente de correlación varía entre -1 y 1, donde 1 indica una correlación positiva perfecta, 0 indica la ausencia de correlación y -1 indica una correlación negativa perfecta.
