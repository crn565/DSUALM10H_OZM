# 8- METRICS

The rapid expansion of NILM and the development of different algorithms have made it essential to provide performance evaluation through performance metrics. Evaluation metrics compare the disaggregation results (predictions) of trained models with validation set data (actual measurements of each process). NILMTK includes calculation of evaluation metrics using MeterGroup for validating results against the validation set.

Let's now analyze different metrics in our dataset such as FEAC, F1, EAE, MNEAP and RMSE which we summarize in the following table:

## F1 Metric

This metric measures the accuracy or assertiveness of the trained model by calculating the harmonic mean using the Precision and Recall metrics. These metrics are calculated internally by NILMTK using the confusion matrix, see the equation in the following figure.

![metricas](metricas.jpg)

Precision is understood as the fraction of time slices where a process was correctly predicted to be active, but was actually off. Conversely, Recall refers to the fraction of time slices where the device was correctly predicted to be on and was actually on.

## EAE Metric

The equation in the figure represents the difference between the total assigned energy yˆt(n) and the total consumed energy yt(n) by process n in kWh

![metricas](metricas.jpg)

```python
round(pd.DataFrame(metrics.loc['EAE']).transpose(),2)
```

```python
predictions_Devices=elec_CO_1s_first.submeters()
e1=predictions_Devices[10].total_energy()[0]
e2=valid_Devices[10].total_energy()[0]
eae=round(abs(e1-e2),2)

pd.DataFrame([e1,e2,eae], columns=['kWh'], index=['Energy_pred','Energy_GT','EAE'])
```

## MNEAP Metric

This indicator, represented by the equation in the image, consists of the difference between the assigned power yˆt(n) and the actual power yt(n) of each process, for each time interval t. After summing all power differences, the result is normalized by the total energy consumption of the process.


![metricas](metricas.jpg)

```python
round(pd.DataFrame(metrics.loc['MNEAP']).transpose(),2)
```

## RMSE Metric

This metric from the equation in the figure is widely used in comparing different datasets. It consists of the root mean square error of the difference between the assigned power yˆt(n) and the actual power yt(n) of appliances, for each time interval t.


![metricas](metricas.jpg)

```python
round(pd.DataFrame(metrics.loc['RMSE']).transpose(),1)
```

## Results of All Models

```python
from nilmtk.metrics import fraction_energy_assigned_correctly, f1_score, error_in_assigned_energy
from nilmtk.metrics import mean_normalized_error_power, rms_error_power 



periods = {'1s':1, '90s':90, '5min':60*5, '10min':60*10}


dir_predictions = data_dir +'NILM/Results/'
m = {}
classifiers = ['CO', 'FHMM']
methods = ['first', 'mean', 'median']
index = list(labels['Meters'][1:12])

for p_name, sp in periods.items():
    met = []
    for method in methods:
        for clf_name in classifiers:
            print('\nCalculando Metricas: '+ clf_name + '_' + p_name + '_' + method)  
           
            #Datos de Dispositivos: predicción
            input_dir = dir_predictions + clf_name + '_' + p_name + '_' + method +'_pred.h5'
            dataset = DataSet(input_dir)
            predictions=dataset.buildings[1].elec.submeters()       
            
            #Metricas
            f1 = {};
            eae = {};
            feac = {};
            mneap = {};
            rmse = {};
            e=[];  
            
            
            F1 = f1_score(predictions, valid_Devices)
            f1['F1'] = F1*100
            EAE = error_in_assigned_energy(predictions, valid_Devices)
            
            for x in range(len(EAE)): e.append(EAE.iloc[x][0])
            eae['EAE'] = e
            #da error para FEAC
            #feac['FEAC'] = 100*np.ones(6)*(fraction_energy_assigned_correctly(predictions, valid_Devices))
            mneap['MNEAP'] = mean_normalized_error_power(predictions, valid_Devices)
            rmse['RMSE'] = rms_error_power(predictions, valid_Devices)
            met.append(pd.concat([pd.DataFrame(feac, index=valid_Devices.instance()),
                                  pd.DataFrame(f1),
                                  pd.DataFrame(eae, index=valid_Devices.instance()),
                                  pd.DataFrame(mneap),
                                  pd.DataFrame(rmse)],
                                 axis=1).set_index(pd.Index(index)))
            dataset.store.close()        
    m[p_name] = met

##Guardamos resultados
f = open(data_dir +'NILM/metrics.txt',"w")
f.write( str(m) )
f.close()
print('\n\nDone!')

```
