### Tasa de abandono y tiempo de funcionamiento

#### Tasa de abandono
Sin contar cuando el sistema no registra datos superando el tiempo máximo de muestreo, aunque también se perciben perdidas de datos debido al propio sistema de adquisición, lo cual se le llama abandono o dropout. 

La tasa de abandono (dropout rate) es por tanto  un número entre 0 y 1 que especifica la proporción de muestras faltantes. Una tasa de abandono de 0 significa que no faltan muestras y un valor de 1 significaría que faltan todas las muestras (100%). 

#### Tiempo de funcionamiento
El tiempo de funcionamiento calculado estará condicionado al parámetro max_sample_period de cada medidor especificado en los metadatos. Si no se han registrado datos pasado  ese tiempo, se considera el medidor como apagado.

```python
a=[]; b=[]; c=[]
for meter, name in Meters.items():
    a.append(name)
    b.append(str(meter.uptime())) #tiempo de funcionamiento
    c.append(round(meter.dropout_rate()*100,1)) #tasa de abandono

a.append('Total')
b.append(str(elec.uptime()))
c.append(round(elec.dropout_rate()*100,1))
    
df=pd.DataFrame(data=[b,c], columns=a, index=['Uptime','Dropout Rate']).transpose()
df
```

### correlation_of_sum_of_submeters_with_mains

El valor devuelto "0.99081" en el contexto de NILMTK (Non-Intrusive Load Monitoring Toolkit) indica la correlación entre la suma de los submedidores y la alimentación principal.La correlación entre la suma de los submedidores y la alimentación principal es una medida de qué tan bien los submedidores están capturando el consumo total de energía del sistema.

Un valor de correlación de 1 indica una correlación perfecta, lo que significa que los submedidores están capturando completamente el consumo de energía total. En este caso, el valor de correlación de 0.99081 indica que hay una alta correlación entre la suma de los submedidores y la alimentación principal, lo que sugiere que los submedidores están capturando la mayor parte del consumo de energía total, pero puede haber una pequeña discrepancia.

### proportion_of_energy_submetered

En NILMTK (Non-Intrusive Load Monitoring Toolkit), el valor devuelto "0.959556" para la métrica "proportion_of_energy_submetered" significa la proporción de energía que se ha medido utilizando submedidores en comparación con la energía total consumida en el sistema.

Esta métrica es útil para evaluar el rendimiento de la desagregación de cargas eléctricas realizada por NILMTK. Un valor de 1 indicaría que todos los dispositivos están siendo correctamente medidos y desagregados por los submedidores, lo que significa que se ha capturado la totalidad de la energía consumida. Un valor menor a 1 indica que hay una parte de la energía que no está siendo medida por los submedidores y, por lo tanto, no se ha desagregado correctamente.

En este caso, el valor de "0.959556" sugiere que se ha medido y desagregado aproximadamente el 95.96% de la energía total consumida en el sistema utilizando submedidores. Esto indica que la desagregación de cargas está funcionando eficientemente, pero aún existe una pequeña proporción de energía que no ha sido capturada o desagregada por los submedidores.

### dropout_rates_ignoring_gaps 

En NILMTK (Non-Intrusive Load Monitoring Toolkit), la métrica "dropout_rates_ignoring_gaps" con los valores "min=0.0, mean=0.0, max=0.0" indica la tasa de "dropouts" (valores faltantes) en los datos de carga eléctrica, ignorando los espacios o brechas entre las mediciones.

En el contexto de la monitorización de cargas eléctricas, los "dropouts" se refieren a los valores faltantes o ausentes en los datos registrados. Estos pueden ser causados por interrupciones en la medición, fallas en los dispositivos de medición o cualquier otro problema técnico. Los espacios o brechas en los datos se refieren a los intervalos de tiempo en los que no se registró ninguna medición.

En este caso, los valores "min=0.0, mean=0.0, max=0.0" indican que no se han encontrado "dropouts" en los datos de carga eléctrica o, en otras palabras, no se han registrado valores faltantes. La tasa de "dropouts" es mínima (0.0) tanto en el valor mínimo como en el valor máximo, y la media también es 0.0, lo que sugiere que los datos de carga eléctrica están completos y no hay brechas significativas en las mediciones. Esto es un buen indicativo de la calidad y la integridad de los datos utilizados en el análisis o la desagregación de cargas eléctricas realizada con NILMTK.
